```python
import pandas as pd
import numpy as np
from faker import Faker
import random
from datetime import datetime, timedelta
import numpy as np
```

```python
fake = Faker()
random.seed(7202023)
# Create data generator function
def generate_data(start_year, end_year, start_date, end_date):
    return [fake.date_between(start_date=start_date, end_date=end_date) for _ in range(start_year, end_year)]

# Generate list of condition names, drug names, and procedure names
condition_names = [
    "Respiratory complication", "Cardiac disorder", "Renal disease", "Hypertension",
    "Asthma", "Diabetes", "Influenza", "Pneumonia", "Tuberculosis",
    "Chronic obstructive pulmonary disease", "Pulmonary embolism", "Lung cancer",
    "Bronchitis", "Emphysema", "Cystic fibrosis", "Sleep apnea", "Lung abscess",
    "Coronary artery disease", "Heart failure", "Stroke", "Arrhythmia",
    "Kidney stones", "Urinary tract infection", "Chronic kidney disease", "Glomerulonephritis",
    "Diabetic nephropathy", "Hypothyroidism", "Hyperthyroidism", "Obesity", "Gastritis",
    "Peptic ulcer disease", "Irritable bowel syndrome", "Crohn's disease", "Ulcerative colitis",
    "Gastroesophageal reflux disease (GERD)", "Gallstones", "Pancreatitis", "Anemia", "Osteoporosis",
    "Rheumatoid arthritis", "Osteoarthritis", "Lupus", "Migraine", "Epilepsy"
]

drug_names = [
    "Amoxicillin", "Azithromycin", "Doxycycline", "Cephalexin", "Ciprofloxacin",
    "Clindamycin", "Metronidazole", "Nitrofurantoin", "Trimethoprim/sulfamethoxazole",
    "Vancomycin", "Albuterol", "Ipratropium", "Budesonide", "Tiotropium", "Roflumilast",
    "Prednisone", "Methylprednisolone", "Hydrocortisone", "Dexamethasone", "Prednisolone",
    "Lisinopril", "Amlodipine", "Metoprolol", "Atorvastatin", "Simvastatin",
    "Levothyroxine", "Esomeprazole", "Omeprazole", "Metformin", "Glipizide",
    "Gabapentin", "Pregabalin", "Ibuprofen", "Acetaminophen", "Aspirin",
    "Naproxen", "Diazepam", "Lorazepam", "Citalopram", "Sertraline",
    "Fluoxetine", "Venlafaxine", "Amitriptyline", "Ranitidine", "Famotidine",
    "Ondansetron", "Loratadine", "Fexofenadine", "Montelukast", "Salmeterol",
    "Insulin", "Sitagliptin", "Liraglutide", "Rivaroxaban", "Warfarin"
] 

procedure_names = [
    "Bronchoscopy", "Chest X-ray", "CT scan of the chest", "Pulmonary function test",
    "Thoracentesis", "Lung biopsy", "Lung surgery", "Thoracotomy", "Pneumonectomy",
    "Lobectomy", "Wedge resection", "Lung transplant", "Radiation therapy", "Chemotherapy",
    "Targeted therapy", "Immunotherapy", "Palliative care", "Coronary angiography", "Angioplasty",
    "Coronary artery bypass grafting (CABG)", "Electrocardiogram (ECG)", "Cardiac catheterization",
    "Echocardiogram", "Pacemaker implantation", "Cardioversion", "Hemodialysis", "Peritoneal dialysis",
    "Kidney transplant", "Lithotripsy", "Endoscopy", "Colonoscopy", "Gastroscopy",
    "Joint replacement surgery", "Arthroscopy", "Hernia repair", "Cataract surgery", "LASIK eye surgery",
    "Appendectomy", "Cholecystectomy", "Hysterectomy", "Caesarean section", "Tonsillectomy",
    "Vasectomy", "Circumcision", "Hemorrhoidectomy", "Gastric bypass surgery", "Sleeve gastrectomy"
]



# Person table
person_ids = list(range(1, 1501))
years_of_birth = generate_data(1, 1501, "-75y", "-10y")
races =A [random.choice(["Asian", "African", "Caucasian", "Hispanic", "Other"]) for _ in range(1500)]
genders = [random.choice(["Male", "Female"]) for _ in range(1500)]
fips_codes = [random.choice([36005, 36047, 36061, 36081, 36085]) for _ in range(1500)]

person_df = pd.DataFrame({"Person_id": person_ids, "Year_of_birth": years_of_birth, "Race": races, 
                          "Gender": genders, "FIPScode": fips_codes})

# Visit_occurrence table
visit_occurrence_ids = list(range(1, 7001))
visit_start_dates = generate_data(1, 7001, "-8y", "-7m")
visit_end_dates = [date + timedelta(days=random.randint(1, 14)) for date in visit_start_dates] # max 2 weeks stay
visit_types = [random.choice(["ED", "inpatient", "outpatient"]) for _ in range(7000)]

visit_occurrence_df = pd.DataFrame({"Visit_occurrence_id": visit_occurrence_ids, "Person_id": [random.choice(person_ids) for _ in range(7000)], 
                                    "Visit_start_date": visit_start_dates, "Visit_end_date": visit_end_dates, 
                                    "Visit_type": visit_types})
visit_person={}
for i in visit_occurrence_df["Visit_occurrence_id"]:
    visit_person[i]=visit_occurrence_df.loc[i-1,"Person_id"]
visit_occurrence_df.loc[visit_occurrence_df["Visit_type"] == "outpatient", "Visit_end_date"] = visit_occurrence_df["Visit_start_date"]
# Condition_occurrence table
condition_occurrence_ids = list(range(1, 7001))
condition_start_dates = generate_data(1, 7001, "-8y", "-7m")
condition_end_dates = [date + timedelta(days=random.randint(1, 14)) for date in condition_start_dates] # max 2 weeks condition
conditions = [random.choice(condition_names) for _ in range(7000)]
condition_occurrence_df = pd.DataFrame({"Condition_occurrence_id": condition_occurrence_ids,
                                        "Visit_occurrence_id": [random.choice(visit_occurrence_ids) for _ in range(7000)],
                                        "Person_id": [random.choice(person_ids) for _ in range(7000)], 
                                        "Condition_start_date": condition_start_dates, 
                                        "Condition_end_date": condition_end_dates, 
                                        "Condition_name": conditions})
# Drug_exposure table
drug_exposure_ids = list(range(1, 7001))
drug_exposure_start_dates = generate_data(1, 7001, "-8y", "-7m")
drug_exposure_end_dates = [date + timedelta(days=random.randint(1, 14)) for date in drug_exposure_start_dates] # max 2 weeks drug exposure
drugs = [random.choice(drug_names) for _ in range(7000)]
drug_exposure_df = pd.DataFrame({"Drug_exposure_id": drug_exposure_ids, 
                                 "Person_id": [random.choice(person_ids) for _ in range(7000)],
                                 "Visit_occurrence_id": [random.choice(visit_occurrence_ids) for _ in range(7000)],
                                 "Drug_exposure_start_date": drug_exposure_start_dates, 
                                 "Drug_exposure_end_date": drug_exposure_end_dates, 
                                 "Drug_name": drugs})

# Procedure_occurrence table
procedure_occurrence_ids = list(range(1, 7001))
procedure_dates = generate_data(1, 7001, "-8y", "-7m")
procedures = [random.choice(procedure_names) for _ in range(7000)]
procedure_occurrence_df = pd.DataFrame({"Procedure_occurrence_id": procedure_occurrence_ids, 
                                        "Person_id": [random.choice(person_ids) for _ in range(7000)],
                                        "Visit_occurrence_id":[random.choice(visit_occurrence_ids) for _ in range(7000)],
                                        "Procedure_date": procedure_dates, 
                                        "Procedure_name": procedures})
```

```python
respiratory_related_procedures = [
    "Bronchoscopy", "Chest X-ray", "CT scan of the chest", "Pulmonary function test",
    "Thoracentesis", "Lung biopsy", "Lung surgery", "Thoracotomy", "Pneumonectomy",
    "Lobectomy", "Wedge resection", "Lung transplant", "Radiation therapy", "Chemotherapy",
    "Targeted therapy", "Immunotherapy", "Palliative care"
]
procedure_occurrence_df["respiratory_related_procedure"] = procedure_occurrence_df["Procedure_name"].apply(
    lambda x: 1 if x in respiratory_related_procedures else 0
)
procedure_occurrence_df["Person_id"] = procedure_occurrence_df["Visit_occurrence_id"].map(visit_person)
```

```python
respiratory_related_conditions = [
    "Respiratory complication", "Asthma", "Influenza", "Pneumonia", "Tuberculosis",
    "Chronic obstructive pulmonary disease", "Pulmonary embolism", "Lung cancer",
    "Bronchitis", "Emphysema", "Cystic fibrosis", "Sleep apnea", "Lung abscess"
]
condition_occurrence_df["respiratory_related_condition"] = condition_occurrence_df["Condition_name"].apply(
    lambda x: 1 if x in respiratory_related_conditions else 0
)
condition_occurrence_df["Person_id"] = condition_occurrence_df["Visit_occurrence_id"].map(visit_person)
```

```python
drug_names_respiratory = [
    "Albuterol", "Ipratropium", "Budesonide", "Tiotropium", "Roflumilast",
    "Prednisone", "Methylprednisolone", "Hydrocortisone", "Dexamethasone"
]
drug_exposure_df["respiratory_related_drug"] = drug_exposure_df["Drug_name"].apply(
    lambda x: 1 if x in drug_names_respiratory else 0
)
drug_exposure_df["Person_id"] = drug_exposure_df["Visit_occurrence_id"].map(visit_person)
```
